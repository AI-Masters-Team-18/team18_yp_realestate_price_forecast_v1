```python
import pandas as pd
from parse_detailed_cian_page import parse_page
from tqdm.notebook import tqdm

df = pd.read_csv("./cianparser/united_filtered.csv", sep=';')
first_thousand_rows = df.head(2)
first_thousand_rows_filtered = first_thousand_rows[['url','underground']]
first_thousand_rows_filtered
```

```python
processed_df = first_thousand_rows_filtered.apply(parse_page, axis=1) #получим Series объек как результат выполнения функции apply
```

```python
processed_df
```

```python
parsed_df = pd.DataFrame.from_records(processed_df)
parsed_df
```

```python
pd.set_option('display.max_colwidth', None)
result_df = pd.concat([df, parsed_df], axis=1)  #объединяем по индксу
print(result_df.head(2).to_string())
```

```python
result_df.to_csv('output.csv', index=False)
```

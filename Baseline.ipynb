Заменим значения "5+" на 5 - таких значений в датасете всего 300, поэтому можно пренебречь количеством комнат в таком случае
Заменим "free" на 0.5 соответсвенно, чтобы было отдельное значение для свободной отделки

```python
def transform_rooms(value):
    if value == '5+':
        return 5
    elif value == 'free':
        return 0.5
    else:
        return int(value)

df['Кол-во комнат'] = df['Кол-во комнат'].apply(transform_rooms)
```

Обучим пока модель на вещественных признаках

```python
numeric_dataset = df.copy()
numeric_dataset = numeric_dataset.select_dtypes(include=['number'])
y = numeric_dataset['Цена']
X = numeric_dataset.drop('Цена',axis=1)
X.shape
```

Предлагаю использовать для оценки качества модели R^2 & RMSE

R^2 поможет в понимании, насколько дальнейшие модели лучше по сравнению с бейзлайном, в то время как RMSE даст понятно интерпретируемый диапазон ошибок в той же размерности (рубли а не рубли^2)

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)

print(f"Train dataset shape: {X_train.shape}")
print(f"Train dataset shape: {y_train.shape}")
print(f"Test dataset shape: {X_test.shape}")
print(f"Train dataset shape: {y_test.shape}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, root_mean_squared_error as RMSE
from sklearn.preprocessing import StandardScaler
```

Применим масштабирование, обучим модели линейной регрессию и Lasso с L1-регуляризацией

```python
from sklearn.preprocessing import StandardScaler
model = LinearRegression()
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_train = pd.DataFrame(data=X_train)
model.fit(X_train, y_train)

X_test = scaler.transform(X_test) #применяем скейлер без обучения на тестовую выборку
X_test = pd.DataFrame(data=X_test)

y_pred = model.predict(X_test)
y_pred_StandartScaler =y_pred
mse = RMSE(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(model)
print(f"Root Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

Подберем альфа коэффициент для регуляризации в Lasso

```python
from sklearn.linear_model import Lasso, ElasticNet
from sklearn.model_selection import GridSearchCV

Lasso_model = Lasso()
Lasso_param_grid = {'alpha': [10, 100, 1000, 10000, 100000], 'max_iter': [1000, 5000]}
Lasso_grid = GridSearchCV(estimator=Lasso_model, param_grid=Lasso_param_grid, scoring='r2', cv=10, verbose=1, n_jobs=-1) #применяем Lasso в GridSearchCV
Lasso_grid.fit(X_train, y_train)
```

```python
Lasso_best_model = Lasso_grid.best_estimator_
Lasso_y_pred = Lasso_best_model.predict(X_test)
print(Lasso_best_model)
r2_test = r2_score(y_test, Lasso_y_pred)
print("R^2 на тестовых данных:", r2_test)
rmse_test = RMSE(y_test, Lasso_y_pred)
print("RMSE на тестовых данных:", rmse_test)
```

```python
#Pipeline для бейзлайн решения
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from category_encoders import LeaveOneOutEncoder
from sklearn.base import BaseEstimator, TransformerMixin

df = pd.read_excel("df_filtered.xlsx")

X_train, X_test, y_train, y_test = train_test_split(df.drop('Цена', axis=1), df['Цена'], test_size=0.3, random_state=42)

class RoomTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, column):
        self.column = column

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()  # Избегаем изменений оригинального DataFrame
        X[self.column] = X[self.column].replace({'5+': 0.5, 'free': 0})
        return X[[self.column]]  # Возвращаем только обработанную колонку
        
categorical_features = ['Ремонт', 'Материал стен', 'Тип жилья']
        
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(drop='first'), categorical_features), #применяем one-hot encoding на ремонт, материал стен и тип жилья
        ('room_transformer', RoomTransformer(), ['Кол-во комнат']), #применяем трансформер на кол-во комнат
        ('loo_encoder', LeaveOneOutEncoder(sigma=0.4),['Станция']), #применяем на станции счетчик со сглаживанием
    ]
)       

pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('scaler', StandardScaler()),
    ('model', Ridge(alpha=10, max_iter=1000))
])

pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)
```

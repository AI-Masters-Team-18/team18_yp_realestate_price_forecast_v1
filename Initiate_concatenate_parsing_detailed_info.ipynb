```python
#реализация парсера через selenium
import pandas as pd
import time
from fake_useragent import UserAgent
from tqdm.notebook import tqdm
import os
import random
# import requests
# from bs4 import BeautifulSoup
# from random import randint

output_file = 'parsed_data.csv'
#при каждом запуске ячейки будем проверять наличие файла с результатами парсинга
if os.path.exists(output_file):
    parsed_df = pd.read_csv(output_file)
    # Определение уже обработанных URL
    processed_urls = set(parsed_df['url'])
    print(f"Уже обработанных URL: {len(processed_urls)}")
    unprocessed_urls = set(df_filtered['url'])-processed_urls
    print(f"Осталось обработать URL: {len(unprocessed_urls)}")
else:
    parsed_df = pd.DataFrame()
    processed_urls = set()

urls_input = result_df.iloc[1039:] # потом заменим на unprocessed_urls

user_agent = UserAgent() 
tqdm.pandas()
start_time = time.time()

for index, row in tqdm(urls_input.iterrows(), total=urls_input.shape[0]):
    random_choice = random.uniform(2, 5)
    time.sleep(random_choice)
    if row['url'] in processed_urls:
        continue  # Пропускаем уже обработанные URL
    parsed_detailed_info_data = parse_page(row, user_agent) #получим словарь с данными
    if not parsed_detailed_info_data : #если функция вернула пустой словарь
        print(f"url {row['url']}, Данные не были получены")
        continue #пропускаем
    else:
        parsed_detailed_info_data['url'] = row['url']  # Добавляем URL в результаты парсинга
        print(f"url {row['url']}, {parsed_detailed_info_data}")
        # Преобразование типов данных
        if 'living_area' in parsed_detailed_info_data:
            parsed_detailed_info_data['living_area'] = float(parsed_detailed_info_data['living_area'])
        if 'kitchen_area' in parsed_detailed_info_data:
            parsed_detailed_info_data['kitchen_area'] = float(parsed_detailed_info_data['kitchen_area'])
        if 'price_per_meter' in parsed_detailed_info_data:
            parsed_detailed_info_data['price_per_meter'] = float(parsed_detailed_info_data['price_per_meter'])
        if 'mins_to_underground' in parsed_detailed_info_data:
            parsed_detailed_info_data['mins_to_underground'] = float(parsed_detailed_info_data['mins_to_underground']) if parsed_detailed_info_data['mins_to_underground'] else None
        parsed_detailed_info_data = pd.DataFrame([parsed_detailed_info_data])
        unprocessed_urls.discard(row['url']) #удаляем обработанные строки из списка необработанных
        # Если parsed_df пустой, инициализируем его как parsed_detailed_info_data
        if parsed_df.empty:
            parsed_df = parsed_detailed_info_data.copy()
        elif not parsed_detailed_info_data.empty:
            # Иначе добавляем данные через concat
            parsed_df = pd.concat([parsed_df, parsed_detailed_info_data], ignore_index=True)
        parsed_df.to_csv(output_file, index=False)
        #сохраняем в файл обработанный url и детали квартиры, файл каждый проход цикла будет пополняться новой строкой с деталями квартиры и url
end_time = time.time()
print("Время выполнения цикла:", end_time - start_time, "секунд")
# output_file = 'parsed_data.csv'
```

```python
import pandas as pd
import time
from fake_useragent import UserAgent

from tqdm.notebook import tqdm
import os
# import requests
# from bs4 import BeautifulSoup
# from random import randint

test_data = pd.read_csv('./united_filtered.csv', sep=';')
test_urls = pd.DataFrame(test_data[['url','underground']].head(4)) #первые 4 ссылки
user_agent = UserAgent() 
output_file_selenium = 'parsed_data_selenium_test.csv'

tqdm.pandas()
start_time = time.time()

parsed_df = pd.DataFrame()
for index, row in tqdm(test_urls.iterrows(), total=test_urls.shape[0]):
    parsed_data = parse_page(row, user_agent) #получим словарь с данными
    print(f"url {row['url']}, {parsed_data}")
    parsed_df = pd.concat([parsed_df, pd.DataFrame([parsed_data])], ignore_index=True)
    parsed_df.to_csv(output_file_selenium, index=False)
# with open("response.html", "w", encoding="utf-8") as file:
#     file.write(html_code)
    
end_time = time.time()
print("Время выполнения цикла:", end_time - start_time, "секунд")

# output_file = 'parsed_data.csv'

# #при каждом запуске ячейки будем проверять наличие файла с результатами парсинга
# if os.path.exists(output_file):
#     parsed_df = pd.read_csv(output_file)
#     # Определение уже обработанных URL
#     processed_urls = set(parsed_df['url'])
# else:
#     parsed_df = pd.DataFrame()
#     processed_urls = set()
    
# # processed_df = first_rows_to_parse_filtered.apply(parse_page, axis=1) #получим Series объек как результат выполнения функции apply
# for index, row in tqdm(first_rows_to_parse_filtered.iterrows(), total=first_rows_to_parse_filtered.shape[0]):
#     if row['url'] in processed_urls:
#         continue  # Пропускаем уже обработанные URL
#     parsed_data = parse_page(row) #получим словарь с данными
#     parsed_data['url'] = row['url']  # Добавляем URL в результаты парсинга
#     parsed_df = pd.concat([parsed_df, pd.DataFrame([parsed_data])])
#     # Сохранение промежуточных результатов
#     parsed_df.to_csv(output_file)
```
